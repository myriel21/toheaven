```python
hp_score.rename(columns={hp_score.columns[0]:'순위',
          hp_score.columns[1]:'국가명',
          hp_score.columns[2]:'점수',
          hp_score.columns[3]:'1인당 소득',
          hp_score.columns[4]:'복지'}, inplace= True)
hp_score.head(10)
```

```python
# 분석에 불필요하다고 판단되는 피처 제거하는 작업
del hp_score['Healthy life expectancy']
del hp_score['Freedom to make life choices']
del hp_score['Generosity']
del hp_score['Perceptions of corruption']
hp_score.head()
```

```python
hp_score.sort_values(by='1인당 소득', ascending=False).head(10)
```

```python
hp_score.sort_values(by='복지', ascending=False).head(10)
```

```python
# os별로 한글처리하는 작업
import platform
from matplotlib import font_manager, rc

if platform.system() == 'Darwin':
    rc('font',family = 'AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font',family=font_name)
else:
    print("Unknown system......")
```

```python
plt.figure(figsize = (6,6))
plt.scatter(hp_score['점수'],hp_score['1인당 소득'],s=50)
plt.xlabel('행복지수')
plt.ylabel('1인당 소득')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.scatter(hp_score['점수'],hp_score['복지'],s=50)
plt.xlabel('행복지수')
plt.ylabel('복지')
plt.grid()
plt.show()
```
