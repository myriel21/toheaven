### 상관계수 검증

#### 상관계수의 값이 통계적으로 유의미한지를 검정하는 것
 - 귀무가설 : 상관관계가 없다.
 - 대립가설 : 상관관계가 있다.
 
#### p-value : 통계의 유인성을 대표하는 지표
 - 귀무가설에 대한 확률값
 - 사회과학분야의 유의수준(알파값 or 임계값): 0.05 즉, 95% 신뢰도
 - 의생명분야의 유의수준:0.01 즉, 99%의 신뢰도
 - p-value >= 0.05 => 대립가설 기각, 귀무가설 채택
 - p-value < 0.05 => 대립가설 채택, 귀무가설 기각

```python
import seaborn as sns

# corr 행렬을 히트맵을 이용하여 시각화하는 작업
hp_score_view = ['순위','점수','1인당 소득', '복지']
sns.set(font_scale = 1.5)
hm = sns.heatmap(corr.values,  # 데이터
                 cbar = True,  # colorbar의 표시여부
                 annot = True,  # 맵상의 셀에 값 표시여부
                 square = True,
                 fmt = '.2f',
                 annot_kws= {'size':15},
                 yticklabels = hp_score_view,
                 xticklabels = hp_score_view)
plt.show()
```
