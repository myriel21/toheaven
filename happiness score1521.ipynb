```python
%matplotlib inline

# 필요한 라이브러리 로딩
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hps_1521 = pd.read_csv('D:/python3/DL/Happiness1521.csv')
```

```python
hps_1521.head()
```

```python
hps_1521.info()
```

```python

pd.pivot_table(hps_1521,index=['Year','Region','Country','Score'])
```

```python
# 특정 value만 표시
pd.pivot_table(hps_1521,index=['Year','Region','Country'],values=['Score'])
```

```python
continent_score =pd.pivot_table(hps_1521,index=['Year','Region'],
               values=['Country','Score'],
               aggfunc=[np.mean],
              fill_value=0)
continent_score
```

```python
A = continent_score[1:9]
A
```

```python
A.sort_values(by=A.columns[0], ascending=False)
```

```python
A.plot.bar(stacked=True)
plt.legend(title='2015Happy Score', loc='center left', bbox_to_anchor=(1,0.5))
plt.show()
```
